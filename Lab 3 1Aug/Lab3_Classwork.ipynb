```python
#Question 1
from math import sqrt
n=int(input('Enter a number: '))
x=int(sqrt(n))
c=0
for i in range(2,x+1):
    if n%i==0:
        c=c+1
        break
if c==1:
    print('The number is not a prime')
else:
    print('This number is prime')
```

```python
#Question 2
#list of marks and grades of 30 students
marks=[34,78,90,14,76,87,54,66,92,100,33,59,23,64,87,88,99,32,67.5,16.5,40.5,47,71,85,65,59,98,100,43,45]
for i in range(0,len(marks)):
    if marks[i]<40:
        print('The marks is',marks[i],'and the grade is F')
    elif marks[i]>=40 and marks[i]<50:
        print('The marks is',marks[i],'and the grade is D')
    elif marks[i]>=50 and marks[i]<60:
        print('The marks is',marks[i],'and the grade is C')
    elif marks[i]>=60 and marks[i]<70:
        print('The marks is',marks[i],'and the grade is B')
    elif marks[i]>=70 and marks[i]<80:
        print('The marks is',marks[i],'and the grade is A')
    elif marks[i]>=80 and marks[i]<90:
        print('The marks is',marks[i],'and the grade is Ex')
    elif marks[i]>=90 and marks[i]<=100:
        print('The marks is',marks[i],'and the grade is TOP')
    else:
        print('Marks not in the range 1-100')
```

```python
#Question 3
def leap(n):
    if n%400==0:
        print('Leap year')
    elif n%4==0 and n%100!=0:
        print('Leap year')
    else:
        print('Not a leap year')
x=int(input('Enter year: '))
leap(x)
```

```python
#Question 8
def fact(n):    #factorial of n
    if n<=1:
        return n
    else :
        return n*fact(n-1)
x=int(input('Enter number'))
fact(x)
```
